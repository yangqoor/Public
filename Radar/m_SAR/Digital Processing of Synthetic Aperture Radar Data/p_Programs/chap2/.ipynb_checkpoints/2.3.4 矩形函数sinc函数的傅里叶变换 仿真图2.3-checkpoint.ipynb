## 矩形和sinc函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rect

Tp = 1e-6
def rect(x):
    return  np.where(np.abs(x)<=0.5,1,0)
    
x = np.linspace(-Tp,Tp,100);
y = rect(x/Tp);
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5*Tp,5*Tp,100);
y = np.sinc(x/Tp);
plt.plot(x,y)
plt.show()
```
