## 2.7.1 N核M点插值表

参照 https://www.freesion.com/article/8626806376/

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
#% 定义N 和 M
N = 2**3;
M = 16;

#% 计算fs
Ts = 1;
fs = 1/Ts;

#% 计算delta 记为d
d = 1/M;

#% kaiser窗函数
beta = 3;
T    = N;
w = np.kaiser(N,beta).reshape(1,N)
# print( "w=",w )
#w    = @(t) besseli(0, beta.*sqrt(1-(2.*t./T).^2) )./besseli(0,beta);

#% 生成数据n和m
n = np.arange(0,N,1).reshape(1,N)
# print( n )
m = np.arange(0,M+1,1).reshape(M+1,1)
# print( m )
a = np.repeat( m*d, N, axis=1 )
# print( a )
b = np.repeat(n, M + 1, axis=0, )
# print( b )
```

```python
#% 生成表格数据kerl
temp = N/2 - 1 + a - b
# print( "temp=",temp)
kerl = np.sinc( temp ) ;#%.* w( temp );
# print( "ini kerl=" )
# print(np.array2string(kerl, formatter={'float_kind':lambda kerl: "%6.4f" % kerl}))
#% 归一化

kerl = np.matmul(np.diag(1./np.sqrt( np.sum(kerl**2, 1) )), kerl)
#print( "kerl=" )
# print(np.array2string(kerl, formatter={'float_kind':lambda kerl: "%6.4f" % kerl}))
#% 加窗
kerl =kerl*(iv(0, beta*np.sqrt(1-(2*temp/T)**2) )/iv(0,beta))
#print(np.array2string(kerl, formatter={'float_kind':lambda kerl: "%6.4f" % kerl}))
#% 显示kerl 与table进行对比
print('插值核为:')
rkerl = kerl[::-1]
print(np.array2string(rkerl, formatter={'float_kind':lambda rkerl: "%6.4f" % rkerl}))

```

```python
#对kerl进行测试
#生成抽样数据
t = np.arange(0,N,1)
gn = np.sin( 2*np.pi*0.25*t)
#生成画图数据
lt = np.arange(0,N-1,0.01)
lgt = np.sin( 2*np.pi*0.25*lt)
#插值
interp_data = np.matmul(kerl,np.transpose(gn))
interp_t = N/2-1+m*d
#% 作图
plt.figure()
plt.plot( t, gn, 'ro')
plt.plot( lt, lgt, 'b')
plt.plot( interp_t,interp_data,"go")
plt.legend(labels=['抽样数据','连续信号数据','插值数据'])
```
