## 2.7.1 加窗sinc函数



```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

kaiser窗加权后

```python
#% 时间序列生成
t = np.arange(-8,8,0.02)

#% 生成未经加权插值核
kernel1 = np.sinc( t );

# 经过加权插值核
kernel2 = np.sinc( t ) * np.kaiser( len(t), 2.5 )

#% 作图
plt.figure()
plt.plot( t, kernel1, 'r', t, kernel2, 'b')
plt.xlabel('时间（采样点）')
plt.ylabel('幅度')
plt.legend(['未经加权','加权'])
plt.title('kaiser窗加权后的sin函数，\\beta=2.5')
```
