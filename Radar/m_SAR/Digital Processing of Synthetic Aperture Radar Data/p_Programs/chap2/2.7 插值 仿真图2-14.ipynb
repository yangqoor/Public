# 《合成孔径雷达成像——算法与实现》2.7.1 sinc函数插值

已经知道g(x)在x_i，i=0，...,k处的值，如何求得其他点的值。

g(x) 可以用如下的阶梯函数近似
$$g(x) = \sum_i g(x_i) rect(x-x_i)$$

用如下的卷积公式

$$g(x) = \sum_i g(x_i) h(x-x_i)$$

其中，h(x)为插值核函数。

如果h(x)设为sinc函数，(时域卷积相当于频域相乘)，故插值可以与sinc核的卷积完成。

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

一维线性卷积

```python
xi = np.arange(8,16,1)
N = len(xi)
x = np.linspace(8, 16, 1000)
K = len(x)
g = np.array([0.2,0.8,0.6,1.5,1.3,-1.0,0.75,1.1])
y = np.zeros(K)
x0 = 11.7
for k in range(K):
    for i in range(N):
        y[k] = y[k]+g[i]*np.sinc(x[k]-xi[i])
        
```

```python
ig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
x1 = np.linspace(-10, 10, 1000)

plt.plot(x1,np.sinc(x1),'b')
plt.show()
```

```python
ig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',x0))
plt.plot(xi,g,'ro')
plt.plot(x,np.sinc(x-x0),'b')
plt.plot(xi,np.sinc(xi-x0),'*')

plt.show()
```

```python
ig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',x0))
plt.plot(xi,g,'ro')
plt.plot(x,y)
plt.show()
```
