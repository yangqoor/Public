# 《合成孔径雷达成像——算法与实现》中2.5.3  奈奎斯特采样率



```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
f0 = 300;  #% 连续正弦波 频率
fs = 800;  #% 抽样频率
Ts = 1/fs;

N  = 8;#%抽样点数N

#% 设置原信号
t1  = np.linspace( 0, (N-1)*Ts, 501 );
ft1 = np.sin( 2*np.pi*f0*t1 );

#% 用800Hz抽样
t2  = np.arange(0,N,1)*Ts;
ft2 = np.sin( 2*np.pi*f0*t2 );


#%% 以400Hz抽样300Hz的信号
#% 参数设置
f0 = 300;  #% 连续正弦波 频率
fs = 400;  #% 抽样频率
Ts = 1/fs;
M  = 5;#% 总抽样点数N

#% 设置原信号
t1_1  = np.linspace( 0, (M-1)*Ts, 201 );
ft1_1 = np.sin( 2*np.pi*f0*t1_1 );

#% 以400Hz抽样
t2_2  = np.arange(0,M,1)*Ts+Ts/3;
ft2_2 = np.sin( 2*np.pi*f0*t2_2 );

```

```python
#画图
plt.figure()
plt.plot( t1_1, ft1_1, 'r', t2_2, ft2_2, 'ko',t2, ft2, 's')
plt.xlabel('t')
plt.legend( ['300Hz连续信号','原信号以400Hz抽样','原信号以800Hz抽样'])
plt.title( '频率为f0的信号以fs抽样（fs<2*f0）')
plt.axis([0,0.01,-1.2,1.2]);
```
