##  《合成孔径雷达成像——算法与实现》中2.5.3  混叠方程

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
fs = 400;
F_continue = np.arange(0,2*fs,0.1);
F_apparent_complex = F_continue-(np.round(F_continue/fs))*fs;
F_apparent_real = np.abs(F_apparent_complex);
plt.figure(figsize=(10,8))
plt.subplot(2,1,1);plt.plot(F_continue,F_apparent_complex);
plt.title('(a)复信号');plt.xlabel('可观测频率');plt.ylabel('可观测频率');
plt.subplot(2,1,2);plt.plot(F_continue,F_apparent_real);
plt.title('(b)实信号');plt.xlabel('可观测频率');plt.ylabel('可观测频率');
plt.tight_layout()
```
