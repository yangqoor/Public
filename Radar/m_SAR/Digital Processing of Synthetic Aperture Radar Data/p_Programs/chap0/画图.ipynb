### 二维曲线

```python
x = np.linspace(0,2*np.pi,100);
y = np.sin(x);
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100);
y = np.sin(x);
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100);
y = np.sin(x);
plt.plot(x,y,':')
plt.plot(x,y2,'r.--')
plt.show()
```

#### 以下例子用于设定标题、xy轴的文字、曲线按颜色说明、设置网格

```python
t = np.linspace(0, 10, 400)
sdata = np.sin(t)
cdata = np.cos(t**2)
plt.plot(t, sdata, label='sin(x)正弦')
plt.plot(t, cdata, label='cos(x^2)余弦')
plt.xlabel("时间(秒)")
plt.ylabel("函数值")
plt.title('显示sin，cos曲线')
plt.grid(True)
plt.tight_layout()
plt.legend(loc='upper right')
plt.show()
```

### 三维曲面图

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y )  # 转换成二维的矩阵坐标
 
 
fig = plt.figure(1, figsize=(12, 8))
zz =  (xx**2 + yy**2)
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.set_top_view()
 
ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```

```python
def f(x, y):
    r =np.sqrt(x**2 + y**2)
    return np.sin(r)*(r)**(-1)

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.view_init(60, 35)
plt.show()
```

```python
from matplotlib import cm
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y)
F = np.sqrt(X**2 + Y**2)
F -= np.min(F)
F /= np.max(F)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,  rstride=1, cstride=1,facecolors=cm.coolwarm(F),shade=False)
plt.show()
```

```python
C = X*Y
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,  rstride=1, cstride=1,facecolors=cm.coolwarm(C),shade=False)
plt.show()
```
