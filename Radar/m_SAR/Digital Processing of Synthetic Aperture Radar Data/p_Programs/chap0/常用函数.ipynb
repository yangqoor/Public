# SAR 常用函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rect

Tp = 1e-6
def rect(x):
    return  np.where(np.abs(x)<=0.5,1,0)
    
x = np.linspace(-Tp,Tp,100);
y = rect(x/Tp);
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5*Tp,5*Tp,100);
y = np.sinc(x/Tp);
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

#defining the x and y arrays
x = np.linspace(0,10, 100)
y = x*np.random.randn(100)**2

#Find peaks
peaks = find_peaks(y, height = 1, threshold = 1, distance = 1)
# print( peaks )
height = peaks[1]['peak_heights'] #list of the heights of the peaks
peak_pos = x[peaks[0]] #list of the peaks positions

#Finding the minima
y2 = y*-1
minima = find_peaks(y2)
min_pos = x[minima[0]] #list of the minima positions
min_height = y2[minima[0]] #list of the mirrored minima heights

#Plotting
fig = plt.figure()
ax = fig.subplots()
ax.plot(x,y)
ax.scatter(peak_pos, height, color = 'r', s = 15, marker = 'D', label = 'Maxima')
ax.scatter(min_pos, min_height*-1, color = 'gold', s = 15, marker = 'X', label = 'Minima')
ax.legend()
ax.grid()
plt.show()
```

## 单位冲激函数

单位样本序列(或者离散时间沖激或单位沖激),用$\delta[n]$来表示,其定义为
$$
\delta[n] = 
\begin{cases}
1, & \quad n=0 \\
0, & \quad n\neq 0
\end{cases}
$$
单位阶跃序列用$\mu[n]$来表示,定义为
$$\mu[n] = 
\begin{cases} 
1, & \quad n\geq 0 \\
0, & \quad n< 0
\end{cases}
$$

```python
n = np.arange(-10,21,1)
hn = np.concatenate((np.zeros([1,10]),np.ones([1,1]),np.zeros([1,20])),axis=1)#,np.zeros([1,20]))
un = np.concatenate((np.zeros([1,10]),np.ones([1,21])),axis=1)

plt.subplot(121)
plt.stem(n,hn.transpose())
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.title('Unit Sample Sequence')
plt.axis([-10,20,0,1.2])

plt.subplot(122)
plt.stem(n,un.transpose())
plt.xlabel('Time index n')
plt.title('Unit Step Sequence')
plt.axis([-10,20,0,1.2])
plt.show()
```
