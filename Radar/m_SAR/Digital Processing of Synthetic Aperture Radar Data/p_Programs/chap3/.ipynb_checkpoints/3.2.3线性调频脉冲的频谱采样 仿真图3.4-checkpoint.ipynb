这里，信号持续时间T = 10us，信号带宽B = 10MHz，改变过采样率（1.4, 1.2, 1.0, 0.8）的取值以探索过采样率对信号频谱的影响。

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
T = 10e-6;                     # % 信号持续时间
B = 10e6;                      # % 信号带宽
K = B/T;                       # % 调频率
ratio = [1.4,1.2,1.0,0.8];
Num = len(ratio);
```

```python
plt.figure()
for ii in range(Num):
    Fs = ratio[ii]*B;     #      % 采样频率
    dt = 1/Fs;              #    % 采样间隔
    N = int(np.ceil(T/dt));         #    % 采样点数
    r = np.arange(N)
    t = (r-N/2)/N*T;     # % 时间轴
    f = (r-N/2)/N*Fs;

    st0 = np.exp(1j*np.pi*K*np.square(t));  #  % 生成信号
    z = np.zeros(N)
    st1 = np.concatenate((st0,z),dtype=complex,axis=None)    # % 补零后的信号，补1倍
    Sf = np.fft.fft(st1);

    n = np.arange(2*N)/2

    plt.subplot(Num,2,2*ii+1)
    plt.plot(t*1e6,np.real(st0))#;axis tight
    plt.ylabel(['\alpha_{os}=',str(ratio[ii])]);
    if(ii == 1):
        plt.title('信号实部');
    if(ii == Num):
        plt.xlabel('时间(\mus)');
    plt.subplot(Num,2,2*ii+2)
    plt.plot(n,np.abs(Sf))#;axis tight
    if(ii ==1):
        plt.title('频谱幅度');
    if(ii == Num):
        plt.xlabel('频率（单元）');
```
