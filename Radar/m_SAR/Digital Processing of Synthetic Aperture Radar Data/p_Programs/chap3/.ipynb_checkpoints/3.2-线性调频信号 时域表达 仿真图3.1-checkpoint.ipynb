## 图3.1 线性调频脉冲的相位及频率



```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
T = 7.24e-6;                #% 信号持续时间
B = 5.8e6;                  #% 信号带宽
K = B/T;                    #% 调频率
ratio = 5;                 #% 过采样率
Fs = ratio*B;              # % 采样频率
dt = 1/Fs;                  #% 采样间隔
N = np.ceil(T/dt);          #   % 采样点数
sr = np.arange(0,N)
t = (sr-N/2)/N*T;      #% 时间轴

st = np.exp(1j*np.pi*K*np.square(t));     #% 生成信号
```

```python
plt.figure()#,set(gcf,'Color','w');
plt.subplot(2,2,1)
plt.plot(t*1e6,np.real(st));
plt.title('(a)信号的实部');plt.xlabel('相对于t_{0}时间(\mus)');plt.ylabel('幅度');
plt.subplot(2,2,2)
plt.plot(t*1e6,np.pi*K*np.square(t));
plt.title('(c)信号的相位');plt.xlabel('相对于t_{0}时间(\mus)');plt.ylabel('弧度');
plt.subplot(2,2,3);
plt.plot(t*1e6,np.imag(st));
plt.title('(b)信号的虚部');plt.xlabel('相对于t_{0}时间(\mus)');plt.ylabel('幅度');
plt.subplot(2,2,4);
plt.plot(t*1e6,K*t*1e-6);
plt.title('(d)信号频率');plt.xlabel('相对于t_{0}时间(\mus)');plt.ylabel('MHz');
```
