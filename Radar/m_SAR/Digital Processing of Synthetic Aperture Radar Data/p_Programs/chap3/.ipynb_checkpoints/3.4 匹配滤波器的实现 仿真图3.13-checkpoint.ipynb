加窗是为了增加目标响应函数的峰值旁边比，可分为时域窗和频域窗。

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
T = 7.24e-6;               # % 信号持续时间
B = 14.2e6;                 # % 信号带宽
K = B/T;                   # % 调频率
ratio = 1.25;                # % 过采样率
Fs = ratio*B;              # % 采样频率
dt = 1/Fs;                 # % 采样间隔
N = int(np.ceil(T/dt));         #    % 采样点数
t = ((np.arange(N))-N/2)/N*T;     # % 时间轴flipud
f = ((np.arange(N))-N/2)/N*Fs; 

st = np.exp(1j*np.pi*K*(t)**2);    # % 生成信号

Sf = np.fft.fft(st);           #    % FFT
Hf = np.exp(1j*np.pi*f**2/K);  #   % 频域匹配滤波器

Out = Sf*Hf;             #  % 频域匹配滤波


ht = np.conj(np.flipud(st));     # % 匹配滤波器
out = np.fft.fftshift(np.fft.ifft(np.fft.fft(st)*np.fft.fft(ht)));
```

```python
window = np.kaiser(N,2.5).transpose(); #  % 时域窗
Window = np.fft.fftshift(window); # % 频域窗
ht_window = window*np.conj(np.flipud(st));

st_window = st*window;    # % 时域加窗
Out_Window = Out*Window;  #  % 频域匹配滤波后加窗
```

```python
fig = plt.figure()#set(gcf,'Color','w');
plt.subplot(2,2,1)
plt.plot(t*1e6,window);
plt.title('时域窗函数');plt.ylabel('幅度');

plt.subplot(2,2,2)
plt.plot(t*1e6,Window)
plt.title('频域窗函数');

plt.subplot(2,2,3)
plt.plot(t*1e6,np.real(st_window))
plt.title('信号实部');plt.xlabel('时间（\mus）');plt.ylabel('幅度');

plt.subplot(2,2,4)
plt.plot(np.real(Out_Window))
plt.title('频谱实部');plt.xlabel('频率（单元）');
fig.tight_layout()

```
