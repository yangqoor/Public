## 附录3A 匹配滤波器的输出 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
#% 设置参数TBP和T
TBP = 100;
T   = 10e-6;

#% 求解B和K
B = TBP/T;
K = B/T;

#% 生成时间序列t
t = np.linspace( -T, T, 2001 );

# % 计算第一部分结果p1 第二部分结果p2 完整结果a
# % 第一部分对T归一化了
p1 = ( T - np.abs(t) )/ T;
p2 = np.sinc( K*t* ( T-np.abs(t) ) );
a  = p1*p2;

#% 图片设置
fig = plt.figure(figsize=(10, 8)) #; set( h, 'position', [100,20,600,650]);
sub_row = 3; sub_col = 1; sub_count = 0;
#% 画图 慢变部分
sub_count = sub_count + 1;  
plt.subplot( sub_row, sub_col, sub_count );
plt.plot( t*1e6, p1 );  plt.ylabel('第一部分');  plt.title('匹配滤波器输出慢变部分')
#% 画图 快变部分
sub_count = sub_count + 1;  
plt.subplot( sub_row, sub_col, sub_count );
plt.plot( t*1e6, p2 );  plt.ylabel('第二部分');  plt.title('匹配滤波器输出快变部分')
#% 画图 慢变部分
sub_count = sub_count + 1;  
plt.subplot( sub_row, sub_col, sub_count );
plt.plot( t*1e6, a ); plt.ylabel('两部分'); plt.title('匹配滤波器输出'); plt.xlabel('时间（us）')
plt.tight_layout()
```

## 匹配滤波器的输出 

```python
#% 生成时间序列t
plt.figure(figsize=(10,4))
t = np.linspace( -T/10, T/10, 201 );

p1 = ( T - np.abs(t) )/ T;
p2 = np.sinc( K*t* ( T-np.abs(t) ) );
a  = p1*p2;
plt.plot( t*1e6, a ); plt.ylabel('幅度'); plt.title('匹配滤波器输出'); plt.xlabel('时间（us）')
b  = t*0
plt.plot( t*1e6, b );
plt.tight_layout()
```
