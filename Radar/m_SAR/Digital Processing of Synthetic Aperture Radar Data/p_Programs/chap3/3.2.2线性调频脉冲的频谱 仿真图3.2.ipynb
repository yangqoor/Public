### 时域FFT仿真

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft,fftshift
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体 
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题 
```

```python
T = 7.24e-6;               # % 信号持续时间
B = 99.43e6;               # % 信号带宽
K = B/T;                   # % 调频率
ratio = 1.25;              # % 过采样率
Fs = ratio*B;              # % 采样频率
dt = 1/Fs;                 # % 采样间隔
N = int(np.ceil(T/dt));            # % 采样点数

t = np.arange(0,N,1)
t = (t-N/2)/N*T;     # % 时间轴

st = np.exp(1j*np.pi*K*t**2);    # % 生成信号
Sf = fftshift(fft(fftshift(st))); #% FFT
# Sf = fft(st); #% FFT

tt = np.linspace(-0.5,0.5,N);
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2,2,1)
plt.plot(tt,np.real(Sf));plt.axis([-0.1,0.1,-50,50]);
plt.title('(a)频谱实部');plt.ylabel('幅度');
plt.subplot(2,2,2)
plt.plot(tt,np.abs(Sf));
plt.title('(b)频谱幅度');plt.ylabel('幅度');
plt.subplot(2,2,3);
plt.plot(tt,np.imag(Sf));plt.axis([-0.1,0.1,-50,50]);
plt.title('(c)频谱虚部');plt.xlabel('频率（归一化后）');plt.ylabel('幅度');
plt.subplot(2,2,4);
plt.plot(tt,np.unwrap(np.angle(Sf)));
plt.title('(d)频谱相位');plt.xlabel('频率（归一化后）');plt.ylabel('相位（弧度）');
plt.tight_layout()
```

### 频域仿真

按照驻定相位原理，信号 $s(t) = rect(\frac{t}{T}) exp\{j\pi K t^2\}$ 的傅里叶变换 
$G(f) = C_1 rect(\frac{f}{KT}) exp\{-j\pi \frac{t^2}{K}\}$，其中 $C_1=\sqrt{\frac{2\pi}{|\phi^{''}(t_s)|}} = \sqrt{\frac{1}{K}}$

```python
#
T = 10e-6;#时宽
B = 30e6;#带宽
ratio = 2;              # % 过采样率
Fs = ratio*B;              # % 采样频率
K = B/T;#调频率
#print(K)
Ts = 1/Fs
N =  int(round( T / Ts )) #采样点数
#print( N )
t = np.linspace( -T/2 , T/2 , N) #%在[-Tp/2,Tp/2]选取采样点
dt = T/N

# 频谱
freq = np.linspace(-Fs/2,Fs/2,N);#频域采样
Gf = np.exp( -1j * np.pi * freq**2 /K ) * np.where( np.abs(freq)/(K*T)<0.5,1,0);
Gf = Gf/np.sqrt(K)

st = np.exp( 1j * np.pi * K * t**2 ) * np.where( np.abs(t/T)<0.5,1,0); 
f_chirp= K * t; #%信号频率
phase_chirp = np.pi * K * t**2;#信号相位
FFTresult = fftshift(fft(st))*dt
```

做一下比较：s(t)做FFT的结果，驻定相位原理得到的G(f)

```python
plt.figure(figsize=(10, 8))
plt.subplot(2,2,1)
plt.plot(t*1e6,np.real(Gf));plt.axis([-1,1,-1e-6,1e-6]);
plt.title('(a)频谱实部');plt.ylabel('幅度');
plt.subplot(2,2,2)
line1=plt.plot(freq/1e6,np.abs(Gf));
line2=plt.plot(freq/1e6,np.abs(FFTresult));
plt.title('(b)频谱幅度');plt.ylabel('幅度');
plt.legend(labels=['近似频谱','实际频谱'], loc='best')
plt.subplot(2,2,3);
plt.plot(t*1e6,np.imag(Gf));plt.axis([-1,1,-1e-6,1e-6]);
plt.title('(c)频谱虚部');plt.xlabel('频率（归一化后）');plt.ylabel('幅度');
plt.subplot(2,2,4);
plt.plot(freq/1e6,-np.pi*freq**2/K);
plt.title('(d)频谱相位');plt.xlabel('频率（归一化后）');plt.ylabel('相位（弧度）');
plt.tight_layout()
```
